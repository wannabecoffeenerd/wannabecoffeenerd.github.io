```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

matplotlib.rcParams.update({'text.usetex': True})
%config InlineBackend.figure_formats = ['svg']
%matplotlib inline
```

```python
df = pd.read_csv('shot_data.csv')
df['Brew Ratio'] = df['Brew Rate'].apply(func = lambda x : float(x.split('/')[1]))
df
```

```python
df0 = df[df['Notes'] == 'No porcupress']
df1 = df[df['Notes'] == 'With porcupress']
```

```python
df0_compare = df0[df0['Grind Setting'] == 5.0]
df1_compare = df1[df1['Grind Setting'] == 5.0]

n0_compare = len(df0_compare)
n1_compare = len(df1_compare)
```

```python
fig, ax = plt.subplots()
fig.patch.set_facecolor('white')
plt.plot(np.zeros(n0_compare), df0_compare['Extraction Yield %'], 'o', color='k', markersize='8', mfc='none')
plt.plot(np.ones(n1_compare), df1_compare['Extraction Yield %'], 's', color='r', markersize='8', mfc='none')
plt.xticks([0, 1], ['Without PorcuPress', 'With PorcuPress'])
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)
plt.xlim((-0.5,1.5))
plt.ylabel('Extraction Yield (\%)', fontsize=16)
plt.tight_layout()
plt.savefig('porcupress_extraction_yield.png', dpi=300, transparent=False)
plt.show()
```

```python
fig, ax = plt.subplots()
fig.patch.set_facecolor('white')
plt.plot(np.zeros(n0_compare), df0_compare['Time'], 'o', color='k', markersize='8', mfc='none')
plt.plot(np.ones(n1_compare), df1_compare['Time'], 's', color='r', markersize='8', mfc='none')
plt.xticks([0, 1], ['Without PorcuPress', 'With PorcuPress'])
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)
plt.xlim((-0.5,1.5))
plt.ylabel('Shot Time (s)', fontsize=16)
plt.tight_layout()
plt.savefig('porcupress_shot_time.png', dpi=300, transparent=False)
plt.show()
```
