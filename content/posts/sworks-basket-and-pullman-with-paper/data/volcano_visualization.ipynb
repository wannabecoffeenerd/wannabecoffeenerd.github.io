```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

matplotlib.rcParams.update({'text.usetex': True})
%config InlineBackend.figure_formats = ['svg']
%matplotlib inline
```

```python
def read_shot_data(filename):
    micros_per_tick = 75.
    df = pd.read_csv(filename)
    df['Ratio'] = df['Yield (g)']/df['Dose (g)']
    df['Burr gap'] = df['Grind']*micros_per_tick
    return df

df0 = read_shot_data('sworksdesign.csv')
df1 = read_shot_data('pullman_paper.csv')
df2 = read_shot_data('wafo_soe_spirit.csv')
df3 = read_shot_data('wafo_soe_spirit_new.csv')
df4 = read_shot_data('pullman.csv')

df1
```

```python
fig, ax = plt.subplots()
fig.patch.set_facecolor('white')
plt.plot(df0['Burr gap'], df0['EY (%)'], 'o', color='k', markersize='8', mfc='none', label=r'Sworksdesign')
plt.plot(df1['Burr gap'], df1['EY (%)'], 's', color='r', markersize='8', mfc='none', label=r'Pullman + paper')
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)
plt.xlabel(r'Burr Gap ($\mu$m)', fontsize=16)
plt.ylabel('Extraction Yield (\%)', fontsize=16)
plt.legend(fontsize=14)
plt.xlim(0,100)
#plt.ylim(23,29)
plt.tight_layout()
plt.savefig('sworksdesign_pullman_paper_extraction_yield.png', dpi=300, transparent=False)
plt.show()
```

```python
fig, ax = plt.subplots()
fig.patch.set_facecolor('white')
plt.plot(df0['Burr gap'], df0['Shot Time (s)'], 'o', color='k', markersize='8', mfc='none', label=r'Sworksdesign')
plt.plot(df1['Burr gap'], df1['Shot Time (s)'], 's', color='r', markersize='8', mfc='none', label=r'Pullman + paper')
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)
plt.xlabel(r'Burr Gap ($\mu$m)', fontsize=16)
plt.ylabel('Shot Time (s)', fontsize=16)
plt.legend(fontsize=14)
plt.tight_layout()
plt.savefig('sworksdesign_pullman_paper_shot_time.png', dpi=300, transparent=False)
plt.show()
```
