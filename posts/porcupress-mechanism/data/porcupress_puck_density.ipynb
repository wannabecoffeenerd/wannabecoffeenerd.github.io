```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

matplotlib.rcParams.update({'text.usetex': True})
%config InlineBackend.figure_formats = ['svg']
%matplotlib inline
```

```python
wo_pp = pd.read_csv('no_porcupress.csv', skiprows=1)
n_wo_pp = len(wo_pp)

wi_pp = pd.read_csv('with_porcupress.csv', skiprows=1)
n_wi_pp = len(wi_pp)
```

```python
fig, ax = plt.subplots()
fig.patch.set_facecolor('white')
plt.plot(np.zeros(n_wo_pp), wo_pp['Thickness (mm)'], 'o', color='k', markersize='8', mfc='none')
plt.plot(np.ones(n_wo_pp), wi_pp['Thickness (mm)'], 's', color='r', markersize='8', mfc='none')
plt.xticks([0, 1], ['Without PorcuPress', 'With PorcuPress'])
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)
plt.xlim((-0.5,1.5))
plt.ylabel('Puck Thickness (mm)', fontsize=16)
plt.tight_layout()
plt.savefig('porcupress_puck_thickness.png', dpi=300, transparent=False)
plt.show()
```

```python
fig, ax = plt.subplots()
fig.patch.set_facecolor('white')
plt.plot(np.zeros(n_wo_pp), wo_pp['Density (g/cm^3)'], 'o', color='k', markersize='8', mfc='none')
plt.plot(np.ones(n_wo_pp), wi_pp['Density (g/cm^3)'], 's', color='r', markersize='8', mfc='none')
plt.xticks([0, 1], ['Without PorcuPress', 'With PorcuPress'])
ax.tick_params(axis = 'both', which = 'major', labelsize = 16)
plt.xlim((-0.5,1.5))
plt.ylabel(r'Puck Density $(\mathrm{g}/\mathrm{cm}^3)$', fontsize=16)
plt.tight_layout()
plt.savefig('porcupress_puck_density.png', dpi=300, transparent=False)
plt.show()
```
